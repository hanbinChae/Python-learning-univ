```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv("data/boston_housing.csv").values

# 데이터 확인 후 지정.
X = dataset[:,:-1]
y = dataset[:,-1]

print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split
xTrain, xTest, yTrain,yTest = train_test_split(X,y,test_size=0.3, random_state=1)
```

```python
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor(min_samples_split=150, random_state=0)
dt.fit(xTrain,yTrain)
#min_samples_split이 작을 경우, overfiting 발생할 수 있음
#min_samples_split이 클 경우, underfiting 발생할 수 있음
```

```python
yTestPred = dt.predict(xTest)
yTrainPred = dt.predict(xTrain)
```

```python
from sklearn.metrics import mean_squared_error
# mean_squared_error : 평균 제곱 오차
# 50일 경우 가장 적합한 일반화
print('*[train mse]:',mean_squared_error(yTrain,yTrainPred)) 
print('*[Test mse]:',mean_squared_error(yTest,yTestPred))
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
plt.figure(figsize=(20,20))
plot_tree(dt)
plt.show()
```

```python
print(dt.tree_.node_count) #dt의 모든 node의 수 
print(dt.tree_.children_left) # 각 node의 좌측 children node index
print(dt.tree_.children_right)# 각 node의 우측 children node index
print(dt.tree_.threshold) 
# 각 node decision에 사용된 임계값 (부등호 기준값)
# ex > x[12] < 10 -> dt.tree_.feature : 12, dt.tree_.threshold : 10
print(dt.tree_.value) 
```
