```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X = np.array([0.5, 1.2, 2.4, 3.8, 5.0, 7.0, 9.0, 9.9,11.5,13.4]).reshape(-1,1)
y = np.array([0.0,0.3,5.5,10.0,11.7,6.0,4.2,2.4,0.9,0.0])

print(X.shape, y.shape)
```

```python
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor(min_samples_split=4, random_state=1)
# min_samples_split = 4 : 4개 까지는 branch로 분리, (4개보다 작을 경우, leaf node)
# randomness?? : 두개의 조건 후, variance가 동일할 때, 주 조건 중 하나를 임의로 결정
dt.fit(X,y)
```

```python
# Decision Tree 시각화
from sklearn.tree import plot_tree
plt.figure(figsize=(8,8))
plot_tree(dt)
plt.show()
```

```python
# model 시각화
plt.scatter(X,y,linewidth=7, alpha=0.5)
xGrid = np.arange(min(X)-1,max(X)+1,0.01).reshape(-1,1)
yGrid = dt.predict(xGrid)
plt.plot(xGrid,yGrid,color='red')
```

```python
# Decisition Tree 결과 분석
print(dt.tree_.node_count) # 전체 노드 수
print(dt.tree_.children_left) #좌측 노드 수
print(dt.tree_.children_right) #우측 노드 수
print(dt.tree_.feature)# 분류 기준이 되는 특징의 index
print(dt.tree_.threshold)# 분류 기준이 되는 특징의 기준값
print(dt.tree_.value) #해당 노드에 속하는 데이터의 predict
```
