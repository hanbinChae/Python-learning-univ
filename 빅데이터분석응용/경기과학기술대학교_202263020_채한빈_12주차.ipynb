```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2,2,20)
y1 = 0.5 *(x-1)**2-1
y2 = -0.7 * (x+0.5)**2+1.5
```

```python
plt.subplot(1,2,1)
plt.plot(x,y1)
plt.plot(x,y2)
```

```python
plt.figure(figsize=(7,3))
plt.subplot(1,2,1)
plt.plot(x,y1)
plt.plot(x,y2)

plt.ylim([-3,3]) #차트의 공백 삭제
plt.xlim([-2,2]) #차트의 공백 삭제

plt.subplot(1,2,2)
plt.plot(x,y2)
plt.ylim([-1,1.5])
plt.xlim([-1,1])
```

```python
plt.figure(figsize=(7,3))
plt.subplot(1,2,1)
plt.plot(x,y1)
plt.plot(x,y2)

plt.ylim([-3,3]) #차트의 공백 삭제
plt.xlim([-2,2]) #차트의 공백 삭제

plt.xticks([-1.5, 0.7,1.3]) #x축에 표시할 숫자 명명

plt.subplot(1,2,2)
plt.plot(x,y2)
plt.ylim([-1,1.5])
plt.xlim([-1,1])
plt.yticks([-0.5,0.2,1], labels=['a','b,','c'])
plt.grid()
```

```python
iris = sns.load_dataset("iris")
iris = iris.drop('species',axis=1)
iris=iris.mean()
iris
```

```python
plt.pie(iris)
plt.show()
```

```python
iris.index
```

```python
plt.pie(iris, labels=iris.index)
plt.show()
```

```python
plt.pie(iris, labels=iris.index, autopct = '%.2f%%')
# autopct = '%f' : pie 차트 안, 백분율 표시
plt.show()
```

```python
plt.pie(iris, labels=iris.index, autopct = '%.2f%%',shadow=True)
# shadow : 그림자 표시, default = False
plt.show()
```

```python
plt.pie(iris, labels=iris.index, autopct = '%.2f%%',shadow=True,
       startangle=90)
plt.show()
```

```python
plt.pie(iris, labels=iris.index, autopct = '%.2f%%',shadow=True,
       startangle=90, explode =[0,0,0.2,0])
```

```python
patches, texts, autotexts = plt.pie(iris, labels=iris.index, autopct = '%.2f%%',shadow=True, startangle=90, explode =[0,0,0.2,0])
plt.setp(autotexts, size=12, color='white')
autotexts[2].set_color('pink')
autotexts[2].set_size(20)

texts[2].set_size(20)
texts[2].set_color('red')
```
