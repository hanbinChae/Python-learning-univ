```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2,2,20)
y1 = 0.5 *(x-1)**2-1
y2 = -0.7 * (x+0.5)**2+1.5
```

```python
plt.subplot(1,2,1)
plt.plot(x,y1)
plt.plot(x,y2)
```

```python
plt.figure(figsize=(7,3))
plt.subplot(1,2,1)
plt.plot(x,y1)
plt.plot(x,y2)

plt.ylim([-3,3]) #차트의 공백 삭제
plt.xlim([-2,2]) #차트의 공백 삭제

plt.subplot(1,2,2)
plt.plot(x,y2)
plt.ylim([-1,1.5])
plt.xlim([-1,1])
```

```python
plt.figure(figsize=(7,3))
plt.subplot(1,2,1)
plt.plot(x,y1)
plt.plot(x,y2)

plt.ylim([-3,3]) #차트의 공백 삭제
plt.xlim([-2,2]) #차트의 공백 삭제

plt.xticks([-1.5, 0.7,1.3]) #x축에 표시할 숫자 명명

plt.subplot(1,2,2)
plt.plot(x,y2)
plt.ylim([-1,1.5])
plt.xlim([-1,1])
plt.yticks([-0.5,0.2,1], labels=['a','b,','c'])
plt.grid()
```

```python
plt.pie(iris, labels=iris.index, autopct = '%.2f%%')
# autopct = '%f' : pie 차트 안, 백분율 표시
plt.show()
```

```python
plt.pie(iris, labels=iris.index, autopct = '%.2f%%',shadow=True)
# shadow : 그림자 표시, default = False
plt.show()
```

```python
plt.pie(iris, labels=iris.index, autopct = '%.2f%%',shadow=True,
       startangle=90)
plt.show()
```

```python
plt.pie(iris, labels=iris.index, autopct = '%.2f%%',shadow=True,
       startangle=90, explode =[0,0,0.2,0])
```

```python
patches, texts, autotexts = plt.pie(iris, labels=iris.index, autopct = '%.2f%%',shadow=True, startangle=90, explode =[0,0,0.2,0])
plt.setp(autotexts, size=12, color='white')
autotexts[2].set_color('pink')
autotexts[2].set_size(20)

texts[2].set_size(20)
texts[2].set_color('red')
```

# 2차시

```python
# 극 축에 선차트 그리기
r = np.arange(0,2,0.01)
theta = 2*np.pi*r

plt.figure(figsize=(5,5))

ax = plt.subplot(111, projection = 'polar')
ax.plot(theta,r)
ax.set_rmax(2)
ax.set_rticks([0,0.5,1,1.5,2])
ax.set_rlabel_position(-30)
```

```python
N = 100
r = 2*np.random.rand(N)
theta = 2*np.pi * np.random.rand(N)
area = 200*np.pi * r**2

plt.figure(figsize=(5,5))
ax = plt.subplot(111,projection='polar')
ax.scatter(theta,r, s=area, c=theta, cmap='hsv')

ax.set_rmax(3)
```

```python
N = 20
r = np.random.rand(N) * 10
theta = 2*np.pi*np.random.rand(N)

plt.figure(figsize=(5,5))
ax = plt.subplot(111,projection='polar')
ax.bar(theta,r,alpha=0.3, color='green')
```

```python
def func(x):
    return (x-3)*(x-5)*(x-7)+85

x = np.linspace(0,10)
x
```

```python
fig = plt.figure(figsize=(6,4))
ax = fig.add_axes([0,0,1,1])
ax.plot( x,func(x),color='red', linewidth=2)
a,b = 2,9
ix = np.linspace(a,b)
iy = func(ix)
ixy = [(a,0),*zip(ix,iy),(b,0)] 

from matplotlib.patches import Polygon
poly = Polygon(ixy, facecolor='lightgray',edgecolor='cyan')

ax.add_patch(poly)
plt.ylim([0,200])

ax.text(0.5*(a+b),30,'Integral between 2 and 9',horizontalalignment='center',fontsize=20)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.yticks([])
plt.xticks([2,9],labels=['a','b'],fontsize=20)
plt.xlabel('X', fontsize=15)
plt.ylabel("Y",fontsize=15)
```
