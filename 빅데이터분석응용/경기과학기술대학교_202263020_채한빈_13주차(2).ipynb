```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 100
r = 2*np.random.rand(N)
theta = 2*np.pi * np.random.rand(N)
area = 200*np.pi * r**2

plt.figure(figsize=(5,5))
ax = plt.subplot(111,projection='polar')
ax.scatter(theta,r, s=area, c=theta, cmap='hsv')

ax.set_rmax(3)
```

```python
def func(x):
    return (x-3)*(x-5)*(x-7)+85

x = np.linspace(0,10)

fig = plt.figure(figsize=(6,4))
ax = fig.add_axes([0,0,1,1])
ax.plot( x,func(x),color='green', linewidth=2)
a,b = 2,9
ix = np.linspace(a,b)
iy = func(ix)
ixy = [(a,0),*zip(ix,iy),(b,0)]

from matplotlib.patches import Polygon
poly = Polygon(ixy, facecolor='lightgray',edgecolor='cyan')

ax.add_patch(poly)
plt.ylim([0,200])

ax.text(0.5*(a+b),30,'Integral between 2 and 9',horizontalalignment='center',fontsize=20)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.yticks([])
plt.xticks([2,9],labels=['a','b'],fontsize=20)
plt.xlabel('X', fontsize=15)
plt.ylabel("Y",fontsize=15)
```
