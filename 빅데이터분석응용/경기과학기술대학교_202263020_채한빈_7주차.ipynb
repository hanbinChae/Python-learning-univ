```python
import pandas as pd
```

```python
nba = pd.read_csv("data/nba.csv")
nba.head()
```

```python
nba.sort_values(by = 'Name') #Name 열로 정렬 (A~Z 순서, 오름차순)
```

```python
nba.sort_values(by = 'Name', ascending = False) #Name열을 기준으로 내림차순으로 정렬
```

```python
nba.sort_values(by='Salary',ascending=False)
```

```python
nba.sort_values(by='Salary',ascending=False, na_position='first')
```

```python
nba.sort_values(by = ['Team','Name'])
```

```python
nba.sort_values(by = ['Team','Name'], ascending=[True,False])
```
