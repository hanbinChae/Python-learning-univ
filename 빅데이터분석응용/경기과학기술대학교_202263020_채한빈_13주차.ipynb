```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
tips.info()
```

```python
sns.relplot(x = 'total_bill',y='tip',data=tips, height=3, aspect=2)
```

```python
tips['size'].unique()
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,hue = 'smoker',height = 3, aspect=2)
# hue : 색상 지정
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,
            hue = 'smoker', style='time', size='size',
           height=3,aspect=2)
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,
            hue = 'smoker', style='time', size='size',
           sizes = (15,200), height=3,aspect=2)
```

```python
sns.relplot(x='total_bill',y='tip',data=tips, hue='smoker', col='time', row='sex',
            height=2.5, aspect=1.5)
```

```python
sns.catplot(x='day', y='total_bill',data=tips, height =2.5, aspect=1.5)
# sns.catplot() : 데이터 값에 대하여 도트를 찍어 주는 것
```

```python
sns.catplot(y='day', x='total_bill',data=tips,
           height =2.5, aspect=1.5)
# x,y 순서 바꿔줌
```

```python
sns.catplot(x='day', y='total_bill',data=tips,
           height =2.5, aspect=1.5,
           order=['Fri','Sat','Sun'])
```

```python
sns.catplot(x='day', y='total_bill',data=tips, 
            height=2.5, aspect=1.5, jitter=False)
# jitter : 진동 없이, 일렬로 데이터 정렬하여 출력
```

```python
sns.catplot(x='day', y='total_bill',data=tips, height =2.5, aspect=1.5, kind='swarm')
# kind='swarm' : marker 자체가 겹쳐지지 않고 출력
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
sns.catplot(x='day',y='total_bill',data=tips, kind='swarm', height=3, aspect=1.5, hue='size')
# == sns.swarmplot(x='day',y='total_bill',data=tips, hue='size')
```

```python
# box plot
sns.catplot(x='day',y='total_bill',data=tips, 
            kind='box', height=2.5, aspect=1.5, palette='Paired')
# == sns.boxplot(x='day',y='total_bill',data=tips)
# palette='Paired' : 색상 옵션 지정 가능, 사이트 참조
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.catplot(x='day',y='total_bill',data=tips, 
            kind='box', height=2.5, aspect=1.5, palette='Set3')
```

```python
sns.catplot(x='day',y='total_bill',data=tips, 
            kind='box', height=2.5, aspect=1.5, palette='Set3',
            hue = 'sex' )
```

```python
sns.catplot(x='day',y='total_bill',data=tips, 
            kind='box', height=2.5, aspect=1.5, palette='Set3',
            hue = 'sex' ,col = 'size')
```

```python
sns.catplot(x='day',y='total_bill',data=tips, 
            kind='box', height=2.5, aspect=1.5, palette='Set3',
            hue = 'sex' , dodge=False) 
# dodge=False : boxplot 중첩해서 출력
```
