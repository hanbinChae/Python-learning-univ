```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
tips.info()
```

```python
sns.relplot(x = 'total_bill',y='tip',data=tips, height=3, aspect=2)
```

```python
tips['size'].unique()
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,hue = 'smoker',height = 3, aspect=2)
# hue :  
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,
            hue = 'smoker', style='time', size='size',
           height=3,aspect=2)
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,
            hue = 'smoker', style='time', size='size',
           sizes = (15,200), height=3,aspect=2)
```

```python
sns.relplot(x='total_bill',y='tip',data=tips, hue='smoker', col='time', row='sex',
            height=2.5, aspect=1.5)
```
