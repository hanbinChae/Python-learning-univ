```python
import pandas as pd
```

```python
bigmac = pd.read_csv("data/bigmac.csv",parse_dates=['Date'])
bigmac.head()
```

```python
bigmac.info()
```

```python
bigmac.set_index(keys = ['Date','Country'], inplace=True) 
#멀티인덱스 지정
```

```python
bigmac.index.names
# 인덱스 이름 출력
```

```python
bigmac.index.get_level_values('Date') 
#모든 Date 인덱스 도출
```

```python
bigmac.index.get_level_values(1)
# idx 2에 해당하는 location 목록 반환 
```

```python
bigmac.index.set_names(names = ['Day',"Location"],inplace=True)
# index 이름 변경
bigmac
```

```python
bigmac.index.set_names(level = 'Day', names ='Date',inplace=True)
bigmac
```

```python
bigmac.sort_index()
#인덱스 정렬
```

```python
bigmac.sort_index(ascending=[True,False])
```

```python
bigmac.sort_index(level='Location', ascending=False)
```

```python
bigmac = bigmac.swaplevel()
```
