```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
satis = pd.read_csv("data/satisfaction.csv")
x = satis['id']
y = satis['Satisfaction Rating']
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y, marker='o', markersize=7)
plt.title('Assignment 2')
```

```python
plt.plot(x,y, marker='o', markersize=7)
plt.xticks([0,30,50])
plt.yticks([0,5,10])
plt.title('Assignment 2')
```

```python
plt.plot(x,y, marker='o', markersize=7)
plt.xticks([min(x),np.mean(x),max(x)])
plt.yticks([0,5,10])
plt.title('Assignment 2')
```

```python
satis = pd.read_csv("data/satisfaction.csv")
x = satis['id']
y = satis['Satisfaction Rating']
p = satis['price']
```

```python
fig = plt.figure(figsize=(7,3))

ax1 = fig.add_axes([0,0,1,1]) 
#x,y:(0,0) , x,y :  1,1 

ax2 = fig.add_axes([0.5,0.6,0.4,0.3])
#x,y:0(12%,60%)  , x,y :  40%,30% 

ax1.plot(x,y,color='red')
ax1.set_xlabel('X')
ax1.set_ylabel("Y")
ax1.set_title('Y1 plot')

plt.scatter(x,p,color='blue')
ax2.set_xlabel('X')
ax2.set_ylabel("Y")
ax2.set_title('Satisfaction Rating & price')
```
