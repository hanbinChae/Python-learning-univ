# 2차시

```python
# 극 축에 선차트 그리기
r = np.arange(0,2,0.01)
theta = 2*np.pi*r

plt.figure(figsize=(5,5))

ax = plt.subplot(111, projection = 'polar')
ax.plot(theta,r)
ax.set_rmax(2)
ax.set_rticks([0,0.5,1,1.5,2])
ax.set_rlabel_position(-30)
```

```python
N = 100
r = 2*np.random.rand(N)
theta = 2*np.pi * np.random.rand(N)
area = 200*np.pi * r**2

plt.figure(figsize=(5,5))
ax = plt.subplot(111,projection='polar')
ax.scatter(theta,r, s=area, c=theta, cmap='hsv')

ax.set_rmax(3)
```

```python
N = 20
r = np.random.rand(N) * 10
theta = 2*np.pi*np.random.rand(N)

plt.figure(figsize=(5,5))
ax = plt.subplot(111,projection='polar')
ax.bar(theta,r,alpha=0.3, color='green')
```

```python
def func(x):
    return (x-3)*(x-5)*(x-7)+85

x = np.linspace(0,10)
x
```

```python
fig = plt.figure(figsize=(6,4))
ax = fig.add_axes([0,0,1,1])
ax.plot( x,func(x),color='red', linewidth=2)
a,b = 2,9
ix = np.linspace(a,b)
iy = func(ix)
ixy = [(a,0),*zip(ix,iy),(b,0)] 

from matplotlib.patches import Polygon
poly = Polygon(ixy, facecolor='lightgray',edgecolor='cyan')

ax.add_patch(poly)
plt.ylim([0,200])

ax.text(0.5*(a+b),30,'Integral between 2 and 9',horizontalalignment='center',fontsize=20)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.yticks([])
plt.xticks([2,9],labels=['a','b'],fontsize=20)
plt.xlabel('X', fontsize=15)
plt.ylabel("Y",fontsize=15)
```

# 3차시

```python
df = pd.read_csv("data/daily-temperature.csv",parse_dates=['Date'])
df
```

```python
df.info()
```

```python
df.set_index("Date",inplace=True)
df.head()
```

```python
df.plot()
```

```python
df.plot(style='.',color='black',alpha=0.3)
# style='.' : scatter plot 과 비슷한 형태로 지정
```

```python
temp_group = df.groupby(pd.Grouper(freq='A'))
```

```python
keys = temp_group.groups.keys()
```

```python
temp_group
```

```python
temp_group.get_group("1986-12-31")
# temp_group.groups.keys() : 데이터 확인
# get_group() : 데이터 추출
```

```python
temp_group.get_group('1981-12-31')['Temp'].values
```

```python
years = pd.DataFrame()
for key in keys:
    years[key] = temp_group.get_group(key)['Temp'].values
years
```

```python
plt.figure(figsize=(5,20))
colors = plt.cm.jet(np.linspace(0,1,10))

for i in range(10):
    plt.subplot(10,1,i+1)
    plt.plot(years.iloc[:,i], color=colors[i])
    plt.title(years.columns[i])
    
plt.tight_layout() #차트 위치 자동 조절, (겹치지 않게)
```

```python
years.plot(subplots =True, figsize= (5,20))
```

```python
plt.matshow(years,aspect='auto', cmap='Oranges')
# 각각의 칸의 색상이 값을 의미한다.
```

```python
years.hist(figsize=(15,15), bins=15,color='red', alpha=0.5)
plt.show()
# 각각의 열에 대해 분포 반환
```
