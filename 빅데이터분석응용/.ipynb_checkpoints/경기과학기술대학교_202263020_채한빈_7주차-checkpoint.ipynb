```python
import pandas as pd
```

```python
nba = pd.read_csv("data/nba.csv")
nba.head()
```

```python
nba.sort_values(by = 'Name') #Name 열로 정렬 (A~Z 순서, 오름차순)
```

```python
nba.sort_values(by = 'Name', ascending = False) #Name열을 기준으로 내림차순으로 정렬
```

```python
nba.sort_values(by='Salary',ascending=False)
```

```python
nba.sort_values(by='Salary',ascending=False, na_position='first')
```

```python
nba.sort_values(by = ['Team','Name'])
```

```python
nba.sort_values(by = ['Team','Name'], ascending=[True,False])
```

```python
nba.sort_index(ascending=False) #인덱스 역순으로 정렬
```

```python
nba.sort_values(by='Name', inplace=True)
nba
```

```python
nba.sort_index(inplace=True)
nba
```

```python
nba['Salary'].rank()
```

```python
nba['Salary'].rank(na_option='bottom')
```

```python
nba['Salary'].rank(ascending=False, na_option='bottom').astype('int')
```

```python
df.info()
```

```python
df['Gender'] == 'Male'
```

```python
df[df['Gender'] == 'Male']
```

```python
mask1 = df['Gender'] =='Male'
mask2 = df['Team'] == 'Marketing'
 # 조건식
df[mask1 & mask2]
```

```python
mask1 = df['First Name'] == 'Robert'
mask2 = df['Team'] == 'Client Services'
mask3 = df['Start Date'] > '2016-06-01'

df[mask1 & mask2 | mask3]
```

```python
import pandas as pd
```

```python
df = pd.read_csv("data/staff.csv",parse_dates=['Start Date','Last Login Time'])
df.head()
```

```python
df['Team'].unique()
```

```python
mask1 = df['Team'] == 'Legal'
mask2 = df['Team'] == 'Product'
mask3 = df['Team'] == 'Sales'
```

```python
df[mask1 | mask2 | mask3]
```

```python
mask = df['Team'].isin(['Legal','Product','Sales']) # isin(): 조건이 여러개인 값을 통합
```

```python
df[mask]
```
