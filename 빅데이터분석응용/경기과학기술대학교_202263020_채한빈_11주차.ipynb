# 1차시

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset("iris")
iris = iris.drop('species',axis=1)
iris.head()
```

```python
iris1 = iris.iloc[0]
iris1
```

```python
iris1.plot.pie()
# pie() : 각 비율을 나타낸 차트
```

```python
iris2 = iris.iloc[0:3]
iris2
```

```python
iris2.plot.pie()
# pie() : 값이 많을 경우 subplots = True로 설정해야 한다.
```

```python
iris2.plot.pie(subplots = True)
# pie() : 값이 많을 경우 subplots = True로 설정해야 한다.
```

```python
iris2.T.plot.pie(subplots = True, figsize=(15,15))
# == iris2.transpose()
# 행과 열 변환
```

```python
iris2.T
```

```python
iris2.T.plot.pie(subplots = True, figsize=(15,15),fontsize=15)
```

```python
iris2.T.plot.pie(subplots = True, figsize=(15,15),fontsize=15, autopct="%.2f")
# autopct="%f" : 차트 안에 백분율로 표시
```

```python
iris2.T.plot.pie(subplots = True, figsize=(15,15),fontsize=15, autopct="%.2f%%")
```

```python
pd.plotting.scatter_matrix(iris,figsize=(7,7), diagonal='kde', color='red')
# scatter_matrix : 두 데이터와의 관계 한눈에 확인
#diagonal='kde' : 데이터 분포 형태 지정
```

```python
iris.plot()
```

```python
iris.plot(subplots=True)
# subplots=True : 각각의 열마다 따로 차트 표시
```

```python
iris.plot(subplots=True, sharex= False)
plt.tight_layout()
# plt.tight_layout() : 차트를 겹치지 않게 조정
#sharex= False : x축 공유하지 않음, (차트마다 각각의 x축 보유)
```

```python
iris.plot(subplots=True, layout=(1,4), figsize=(12,3))
#  layout=(1,4) : 차트의 방향 지정
```

```python
iris.plot(subplots=True, layout=(2,2), figsize=(7,7),sharex=False)
```

```python
iris.plot(subplots=True, layout=(2,2), figsize=(7,7),sharex=False, fontsize=12)
```

# 2차시

```python
x = np.linspace(-2,2,20)
x
```

```python
y = 0.5*(x-1)**2 -3
y
```

```python
plt.plot(y)
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y , color='magenta')
```

```python
plt.plot(x,y,color='purple',marker='o')
```

```python
plt.plot(x,y,color='purple',marker='x',markersize=7, linestyle='-.')
plt.title("Line plot", fontsize=20, color='red')
plt.xlabel("x axis",fontsize=15, color='green')
plt.ylabel("y axis",fontsize=15, color='green')
plt.grid(linestyle='--',color='pink') # 격자무늬 표시
```

```python
iris = sns.load_dataset('iris')
iris=iris.drop("species",axis=1)
iris.head()
```

```python
x = iris['petal_length']
y = iris['petal_width']
```

```python
plt.scatter(x,y, linewidth=5)
# linewidth=5 : 각 marker별 크기 지정
```

```python
plt.scatter(x,y, linewidth=5, color = 'green',alpha = 0.5)
plt.title("Scatter plot")
plt.xlabel("petal_length")
plt.ylabel("petal_width")
plt.grid()
```

```python
iris_mean = iris.mean(axis=0)
iris_mean
```

```python
x = iris_mean.index
y = iris_mean
plt.bar(x, y)
```

```python
plt.bar(x, y, width=0.4, color = 'purple', alpha=0.6)
# width = 0~1 : bar의 폭 지정
```
