```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### 1. score1.csv 를 로드하고 score에 저장하고, 하위 3개의 행만 출력하시오.

```python
# 코드 작성
score = pd.read_csv("data/score1.csv")
score.tail(3)
```

### 2. 'name'열을 index열로 지정하시오. (변경사항을 score 데이터프레임에 저장)¶

```python
# 코드 작성
score = score.set_index("name")
```

```python
score
```

### 3. Sarah, Thomas, Ann학생의 math, science, industry 점수를 새로운 데이터프레임 score_part에 저장하시오.

```python
# 코드 작성
score_part=score[['math','science','industry']]
```

```python
score_part
```

### 4. 미술(art)과 음악(music) 점수가 아래와 같을 때, 데이터프레임 score에 'art'과 'music'열을 추가하시오. (변경사항을 score 데이터프레임에 저장)

- art: 모두 83점
- music: 순서대로 90,87,88,79,91,82,78,80

```python
# 코드 작성
score['art']=83
score['music']=[90,87,88,79,91,82,78,80]
```

```python
score
```

### 5. score2.csv를 로드한 후, score 데이터프레임에 연결하시오. (변경사항을 score 데이터프레임에 저장)

```python
# 코드 작성
score2 = pd.read_csv("data/score2.csv")
score2 = score2.set_index('name')
score = pd.concat([score, score2], axis=0, join='outer')
```

```python
score
```

### 6. 누락 데이터를 해당 과목 평균값으로 대체하시오. (변경사항을 score 데이터프레임에 저장)

```python
# 코드 작성
for i in range(7):
    score.iloc[:,i] = np.where(np.isnan(score.iloc[:,i]),np.mean(score.iloc[:,i]),score.iloc[:,i])
```

```python
score
```

### 7. 데이터프레임 score에 'total'열을 추가하고, 각 학생 점수의 합계를 입력하시오. (변경사항을 score 데이터프레임에 저장)

```python
# 코드 작성
score['total']=np.sum(score,axis=1)
```

```python
score
```

### 8. total 값을 기준으로 내림차순 정렬하시오. (변경사항을 score 데이터프레임에 저장)

```python
# 코드 작성
score = score.sort_values('total',ascending=False)
```

```python
score
```

### 9. 'avg' 열을 생성하여 'total' 열 값을 7로 나눈 값을 저장하고, 'total' 열을 삭제하시오. (변경사항을 score 데이터프레임에 저장)

```python
# 코드 작성
score['avg'] = score['total']/7
score = score.drop('total',axis=1)
```

```python
score
```

### 10. 각 교과목의 평균을 막대 차트(bar plot)으로 나타내시오. (가로축에 과목명이 나타나도록 그리고, 막대는 파란색이 아닌 색으로 그리시오)

```python
# 코드 작성
score_mean = np.mean(score[['korean','english','math','science','industry','art','music']])
score_mean.plot.bar(color='red') 
```

### 11. avg 값(평균 점수)의 구간을 3개로 나누어 히스토그램을 그리시오.

```python
# 코드 작성
score['avg'].plot.hist(bins=3)
```
