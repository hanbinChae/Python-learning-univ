# 3차시

```python
chicago = pd.read_csv("data/chicago.csv")
chicago
```

```python
chicago.dropna(how='all',inplace=True) 
#NaN이 포함된 데이터 모두 삭제
```

```python
chicago
```

```python
chicago.info()
```

```python
chicago['Department'].nunique()
#중복 없는 데이터의 개수 확인
```

```python
chicago['Department'] = chicago['Department'].astype('category')
# Department 열의 데이터 타입을 category로 변경
chicago.info()
```

```python
chicago
```

```python
chicago['Name'].str.lower().str.upper()
#.str.lower() : 모든 문자를 소문자로 변경
#.str.upper() : 모든 문자를 대문자로 변경
```

```python
chicago['Position Title'].str.title()
#str.title : 각 단어의 첫번째 문자만 대문자로 변경
```

```python
chicago['Department'].str.len() 
# 각 열의 길이 반환
```

```python
chicago['Department'] = chicago['Department'].str.replace('MGMNT','MANAGEMENT')
# str.replace : 특정 문자열을 지정한 문자열로 변경
```

```python
chicago.head()
```

```python
chicago['Employee Annual Salary'] = chicago['Employee Annual Salary'].str.replace('$','')
```

```python
chicago['Employee Annual Salary'] = chicago['Employee Annual Salary'].astype("float")
```

```python
chicago.info()
```

```python
chicago.head()
```

```python
mask = chicago['Position Title'].str.upper().str.contains("WATER")
# 오류를 방지하기 위해, 대문자 변환 후, 원하는 키워드 contains 함수로 찾기
# contains : 원하는 데이터가 포함되었다면 True로 반환
chicago[mask]
```

```python
mask = chicago['Position Title'].str.startswith("WATER")
# startswith() : ~로 시작하는 데이터 True로 반환
chicago[mask]
```

```python
mask = chicago['Position Title'].str.lower().str.endswith('ist')
# endswith() : ~로 끝나는 데이터를 True로 반환
chicago[mask]
```

```python
'    Hello  '.strip()
#strip() : 공백 삭제
#lstrip(), rstrip() : 각각 왼쪽, 오른쪽 공백 삭제
```

```python
chicago
```

```python
chicago['Name'].str.split(',')
```

```python
FName = chicago['Name'].str.split(',').str.get(1).str.strip().str.split(' ').str.get(0)
# ,를 기준으로 데이터를 분리한 후, 1번 인덱스 추출
```

```python
FName.value_counts()
# value_counts() : 각각의 데이터가 몇번 등장하는지 갯수 확인
```

```python
chicago['First Name'] = FName
```

```python
chicago
```

```python
chicago['Name'].str.split(',', expand=True)
# split, expand=True : 데이터 프레임으로 반환
```

```python
chicago[['L Name','F Name']] = chicago['Name'].str.split(',',expand=True)
chicago
```
