# 3차시

```python
import pandas as pd
```

```python
fortune = pd.read_csv("data/fortune1000.csv",index_col = 'Rank')
fortune.head()
```

```python
sectors = fortune.groupby('Sector')
sectors
```

```python
sectors.get_group('Energy')
```

```python
sectors.max()
```

```python
sectors.sum()
#문자열 자동 제외
```

```python
sectors.mean()
```

```python
sectors['Revenue']
#하나의 열만 선택할 경우 Series group by 형태로 반환
```

```python
sectors['Revenue'].sum()
```

```python
sectors[['Revenue','Profits']].sum()
```

```python
sectors[['Revenue','Profits']].mean()
```

```python
sectors[['Revenue','Profits','Employees']].max()
```

```python
sectors2 = fortune.groupby(['Sector','Industry'])
sectors2.size()
```

```python
list(sectors2.groups.keys())[3]
```

```python
for i in sectors2.groups.keys():
    print(f"*keys: {i}")
    print(sectors2.get_group(i))
```

```python
sectors.agg(['size','sum','mean'])
# agg() : 각 열별로 다른 함수를 적용
```

```python
sectors.agg({'Revenue':['sum','mean'],'Profits':['mean','max'],'Employees':'max'})
# agg() : 각 열별로 다른 함수를 적용
```

```python
result = pd.DataFrame(columns = fortune.columns)
result
```

```python
for key,df in sectors:
    max_rev = df.nlargest(1,"Revenue")
    result = result.append(max_rev)
```
