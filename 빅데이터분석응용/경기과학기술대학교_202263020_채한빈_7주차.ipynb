# 2차시

```python
import pandas as pd
```

```python
df = pd.read_csv("data/staff.csv",parse_dates=['Start Date','Last Login Time'])
df.head()
```

```python
df['Team'].unique()
```

```python
mask1 = df['Team'] == 'Legal'
mask2 = df['Team'] == 'Product'
mask3 = df['Team'] == 'Sales'
```

```python
df[mask1 | mask2 | mask3]
```

```python
mask = df['Team'].isin(['Legal','Product','Sales']) # isin(): 조건이 여러개인 값을 통합
```

```python
df[mask]
```

```python
df['Team'].unique()
```

```python
mask = df['Team'].isnull() #null인 데이터 True
df[mask]
```

```python
mask = df['Team'].notnull() # null인 아닌 데이터 True
df2 = df[mask]
```

```python
df2
```

```python
mask = df['Salary'].between(60000,70000, inclusive = 'both') #특정 범위의 실수값 True로 반환
df[mask]
```

```python
df.sort_values('First Name', inplace=True)
df.head(6)
```

```python
df['First Name'].duplicated() # 연속적으로 발생하는 데이터 True로 반환, (첫번쨰 데이터 : False, 연속적인 데이터 : True)
```

```python
df[df['First Name'].duplicated()]
```

```python
df['First Name'].duplicated(keep = 'last').head(8) #(마지막 데이터 : False, 연속적인 데이터 : True)
```

```python
df[df['First Name'].duplicated(keep = 'last')].head(8) #마지막 데이터 False이므로, 출력되지않음
```

```python
df['First Name'].duplicated(keep=False) #keep : 연속적인 데이터는 모두 True, 한번만 등장한 데이터는 False
```

```python
df[df['First Name'].duplicated(keep=False)] #모두 중복된 데이터만 출력
```

```python
~df['First Name'].duplicated(keep=False)
```

```python
df[~df['First Name'].duplicated(keep=False)]
```

```python
df.drop_duplicates(subset="First Name", keep='first') #처음 행만 남겨두고 연속적인 데이터 삭제
```

```python
df.drop_duplicates(subset="First Name", keep=False)
```

```python
df.drop_duplicates(subset=['First Name','Team'], inplace=True) 
#First Name과 Team이 모두 연속되어 나타날 경우 삭제
```
