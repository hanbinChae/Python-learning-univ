# **원형 연결리스트**

```python
##노드 생성
class Node():
 def __init__(self):
  self.data=None
  self.link=None

node1=Node()
node1.data="수한무"
node1.link=node1 ##원형링크 첫번째 노드생성
print(node1.data, end=' ')
print(node1.link.data)
```

# **원형 연결리스트 삽입**

```python
dataArray = ['수한무','거북이','두루미','삼천갑자','동박삭']
 
node = Node()
node.data = dataArray[0] # 수한무
head = node
node.link=head

current = node
for data in dataArray[1:]:
  node = Node()
  node.data = data
  current.link = node 
  node.link = head
  current = node

# print(head.data,end=' ')
# print(head.link.data,end=' ')
# print(head.link.link.data,end=' ')
# print(head.link.link.link.data,end=' ')
# print(head.link.link.link.link.data,end=' ')
# print(head.link.link.link.link.link.data,end=' ') #원 전체 순환

#연결리스트 내부 확인
p = head
for _ in range(len(dataArray)+1):
  print(p.data,end=' ')
  p = p.link
```

# **원형 연결리스트 맨앞에 데이터 추가**

```python
node = Node()
node.data ='김'
node.link = head

last = head
while last.link != head:
  last = last.link #김->한칸씩 이동

last.link = node
head = node

p = head
for _ in range(len(dataArray)+2):
  print(p.data,end=' ')
  p = p.link
```

# **원형 연결리스트 중간에 데이터 추가**

1. 중간 위치에 노드 삽입 (거북이를 찾아 자라 삽입)-> 두번째 노드부터 삽입하는 로직

```python
# 노드 삽입
# 중간 노드 삽입(거북이를 찾아 자라를 추가) -> 두번째 노드부터 삽입하는 로직
# 1. 거북이 찾기 -> 2. pre,current에 저장하며 찾기. -> 3. 거북이 찾으면 새노드 생성-> 4. 새노드.link
pre = head
current = head
while current.link!=head:
 pre=current
 current=current.link
 #print(pre.data) #첫번째 노드 값
 print("while문장 윗줄: ",current.data) # 두번째 노드 값

 if current.data =="거북이":
 #print("if 문장 안 : ",current.data) # 두번째 노드 값
  node=Node()
  node.data="자라"
  node.link=current
  pre.link=node
  break

 print("while문장 뒷줄: ", current.data) #첫번째 노드 값


if current.link == head:
 #print("last",head.data) #첫번째 노드 값
 #print(head.link.data) # 두번째 노드 값
 node=Node()
 node.data="자라"
 node.link=head
 current.link=node
print(head.data, end=' ') #첫번째 노드 값
print(head.link.data, end=' ') # 두번째 노드 값
print(head.link.link.data, end=' ')
print(head.link.link.link.data, end=' ')
print(head.link.link.link.link.data, end=' ')
print(head.link.link.link.link.link.data, end=' ')
print(head.link.link.link.link.link.link.data, end=' ')
print(head.link.link.link.link.link.link.link.data, end=' ') 
```

# **원형 연결리스트 첫번째 노드 삭제**
1. 첫번째 노드를 current로 지정
2. head를 다음 위치 노드로 지정
3. 마지막 노드 찾아서 한 노드씩 이동하며 검색 -> 마지막노드.link = head
4. current 노드 삭제

```python
# 노드 삭제
# 첫번째 노드 삭제
# 1. current를 head에 위치->2.head 를 삭제할 노드가 가리키는 노드로 이동->3.last노드 찾아서->4.last.
current=head
head=head.link
last=head
while last.link !=current:
 last=last.link

last.link=head
del(current)
print(head.data, end=' ') #첫번째 노드 값
print(head.link.data, end=' ') # 두번째 노드 값
print(head.link.link.data, end=' ')
print(head.link.link.link.data, end=' ')
print(head.link.link.link.link.data, end=' ')
print(head.link.link.link.link.link.data, end=' ')
```

# **원형 연결리스트 원하는 노드 삭제**

```python
# 노드 삭제
# 중간 노드 삭제
# 1. 자라 찾기 -> 2. pre,current에 저장하며 찾기. -> 3. 자라 찾으면 pre.link=current.link로 수정-> 4
pre=head
current=head
while current.link !=head:
 pre=current
 current = current.link
 if current.data =="자라":
  pre.link=current.link
 del(current)
 break
print(head.data, end=' ') #첫번째 노드 값
print(head.link.data, end=' ') # 두번째 노드 값
print(head.link.link.data, end=' ')
print(head.link.link.link.data, end=' ')
print(head.link.link.link.link.data, end=' ')
print(head.link.link.link.link.link.data, end=' ') 
```
