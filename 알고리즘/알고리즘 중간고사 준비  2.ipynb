```python
link_list = ['채'+str(idx) for idx in range(1,6)]

# print(link_list)

class Node():
  def __init__(self):
    self.data = None
    self.link = None

node = Node()
node.data = link_list[0]
head = node

for l in link_list[1:]:
  pre = node
  node = Node()
  node.data=l
  pre.link = node

temp = head
while temp.link:
  print(temp.data,end=' ')
  temp = temp.link
print(temp.data)
```

```python
node = Node()
node.link = head
node.data = '처음'
head = node

temp = head
while temp.link:
  print(temp.data,end=' ')
  temp = temp.link
print(temp.data)
```

```python
#중간 노드 삽입
pre = head
current = head.link

while current.link != None:
  if current.data == '채3':
    node = Node()
    node.link = current
    node.data= '중간채'
    pre.link = node
    break;
  pre = current  
  current = current.link

node = Node()
node.data ='마지막'
current.link = node

temp = head
while temp.link:
  print(temp.data,end=' ')
  temp = temp.link
print(temp.data)
```

```python
# 중간 위치에 삽입
current = head

while current.link != head:
  pre = current
  current = current.link
  if current.data == '노':
    node = Node()
    node.data = '노랑색전'
    pre.link = node
    node.link = current
    break;

temp = head
while temp.link:
  print(temp.data,end=' ')
  temp = temp.link
print(temp.data)
```
