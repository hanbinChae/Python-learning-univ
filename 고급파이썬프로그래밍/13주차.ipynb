클래스 생성 개수

```python
class Rect:
    count = 0
    
    def __init__(self,width,height):
        self.width, self.height = width, height
        Rect.count+=1
        print(f'{Rect.count}번 인스턴스')
        
        
if __name__ == '__main__':
    r1 = Rect(3,5)
    print('r1 (width, height, count) = ',r1.width,r1.height,r1.count)

    r2 = Rect(4,7)
    print('r2 (width, height, count) = ',r2.width,r2.height,r2.count)
```

사각형 속성 처리

```python
class Rect:
    count = 0
    
    def __init__(self,width,height):
        self.width, self.height = width, height
        Rect.count +=1;
        self.rectid = "rectid_"+str(Rect.count)
    
    def get_area(self): # 사각형 넓이 계산
        return (self.width * self.height)
    
    def get_peri(self): #사각형의 둘레를 계산
        return 2 * (self.width + self.height)

r1 = Rect(3,5)
r2 = Rect(4,7)

area = r1.get_area()
peri = r1.get_peri()

print(f"넓이 : {area}\n둘레 : {peri}")
```

**`__repr__ : 정보 표기 메`소드**

```python
class Vect2:
    ndim = 2 #클래스 변수
    
    def __init__(self,x,y): #생성자
        self.x,self.y = x,y
    
    def __repr__(self): #정보 표기 메소드
        return f"Vector[{self.x}, {self.y}]"
    
v1 = Vect2(3,5)
print(f'v1 = {v1}')
print(f'v1(x,y) = {v1.x,v1.y}')
```

```python
class Vect2:
    ndim = 2 #클래스 변수
    
    def __init__(self,x,y): #생성자
        self.x,self.y = x,y
    
    def __repr__(self): #정보 표기 메소드
        return f"Vector[{self.x}, {self.y}]"
    
    def __add__(self,other):
        new = Vect2(0,0)
        new.x = self.x + other.x
        new.y = self.y + other.y
        return new
        
v1 = Vect2(3,5)
v2 = Vect2(7,4)
v3 = v1 + v2

print(v1,v2,v3)
```

캡슐화 (Private)

```python
class Vect3:
    ndim = 2 #클래스 변수
    
    def __init__(self,x,y=0):
        self.x,self.y = x,y
        
    def __repr__(self):
        return f'Vector[{self.x},{self.y}]'
    
    def __add__(self,other):
        __new = Vect3(0,0)
        __new.x = self.x + other.x
        __new.y = self.y + other.x
        return __new
    
v1 = Vect3(3,5)
v2 = Vect3(7)
v3 = v1+v2
print(v1,v2,v3)
print(v2.x, v2.y)
print(v1.__new) # private 결과값을 호출하려고 할 때 에러 발생, (__변수명 : private 멤버 : 다른 클래스의 메소드에서 직접 접근 불가능)
```

상속

```python
class Rect:
    def __init__(self,width,height=-1):
        if height < 0: # 정사각형인 경우
            height = width
        self.width, self.height = width, height
    def __repr__(self): # 메소드 오버라이딩
        return f'Square(width = {self.width},height = {self.height})'
    
    def get_area(self): # 사각형 넓이 계산
        return (self.width * self.height)
    
    def get_peri(self): #사각형의 둘레를 계산
        return 2 * (self.width + self.height)
    
    def isSquare(self): #정사각형인지 확인
        return (self.width == self.height)

class Square(Rect):
    def __repr__(self): # 메소드 오버라이딩
        return f'Square(width = {self.width})'
    
    def isSquare(self):
        return True
    

s1 = Rect(5,5) #정사각형일때
print(s1,", 면적:",s1.get_area())
print(s1.isSquare())

s2 = Square(5) #정사각형일때
print(s1,", 면적:",s1.get_area())
print(s1.isSquare())

s1 = Rect(5,7) #정사각형일때
print(s1,", 면적:",s1.get_area())
print(s1.isSquare())
```
